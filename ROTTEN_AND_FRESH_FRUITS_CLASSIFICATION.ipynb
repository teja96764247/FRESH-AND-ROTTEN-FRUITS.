```python
x=[]
y=[]
for i,j in data:
  x.append(i)
  y.append(j)
```

```python
x1=np.array(x)
y1=np.array(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x1,y1,test_size=0.1,random_state=2)
```

```python
print(x1.shape,x_train.shape,x_test.shape)
```

```python
x_train_scaled=x_train/255
x_test_scaled=x_test/255
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(32,32,3)),
    keras.layers.Dense(64,activation='relu'),
    keras.layers.Dense(128,activation='relu'),
    keras.layers.Dense(6,activation='softmax')
])
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train_scaled,y_train,epochs=50)
```

```python
input_img_path='/content/apple.png'
img_pillow=Image.open(input_img_path)
img_color=img_pillow.convert('RGB')
img_resized=img_color.resize((32,32))
img_arr=np.array(img_resized)
cv2_imshow(img_arr)
img_scaled=img_arr/255
input_image_scaled=np.reshape(img_scaled,[1,32,32,3])
prediction=model.predict(input_image_scaled)
result=np.argmax(prediction)
if(result==0):
  print('Yeah it was a FRESH APPLE')
elif(result==1):
  print('Yeah it was a FRESH BANANA')
elif(result==2):
  print('Yeah it was a FRESH ORANGE')
elif(result==3):
  print('Yup it was ROTTEN APPLE')
elif(result==4):
  print('Yup it was ROTTEN BANANA')
elif(result==5):
  print('Yup it was a ROTTEN ORANGE')
```
